```python
import tensorflow_decision_forests as tdft
```

##Load the training dataset

```python
from google.colab import files
```

```python
import pandas
train_df = pandas.read_csv("/content/penguins.csv")
```

## Convert the pandas dataframe into tensorflow datatset

## the label is our y,what we are plannin to predicct/classify

```python
train_ds = tdft.keras.pd_dataframe_to_tf_dataset(train_df, label="species")
```

## Train the model

```python
model=tdft.keras.RandomForestModel()
model.fit(train_ds)
```

## Interpreting a model

```python
tdft.model_plotter.plot_model_in_colab(model,tree_idx=0)
```
